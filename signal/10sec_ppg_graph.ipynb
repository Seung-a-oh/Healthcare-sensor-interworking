# 10sec graph with CSV

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv('threading_ppg.csv', encoding="UTF-8")
df
```

```python
sec = list(np.arange(0,10, 0.01))

df["sec"] = sec
df
```

```python
ppg = df.loc[:1000,'ppg']
beat = df.loc[:1000,'beat']
sec = df.loc[:1000,'sec']
```

```python
# PPG 그래프

plt.figure(figsize=(100,50))
plt.xlabel('sec',fontsize=20)
plt.ylabel('ppg value(v)',fontsize=20)
plt.plot(sec, ppg, color="r")
# plt.plot(sec, beat)
plt.savefig("./threading_ppg2.png")
```

```python
# BEAT 그래프

plt.figure(figsize=(50,10))
font = {
#     'color':'deeppink',
    'size':20
}
plt.xlabel('sec',fontdict=font)
plt.ylabel('beat value', fontsize=20)
plt.plot(sec, beat)
plt.savefig("./threading_beat1.png")
```

```python
# PPG & BEAT 그래프

plt.figure(figsize=(100,20))
font = {
    'size':20
}
plt.xlabel('sec',fontdict=font)
plt.ylabel('beat value', fontsize=20)
plt.plot(sec, ppg)
plt.plot(sec, beat)
plt.savefig("./threading_ppg and beat.png")
```
